Hiring probability of candidate based on experience and skills Consider three simple rules:

If the experience is low or the skills are low, then the hiring probability will be low. If the experience is medium, then the hiring probability will be medium. If the experience is high OR the skills are high, then the hiring probability will be high.

```python
!pip install scikit-fuzzy
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

```python
#   * Experience and Skill on subjective ranges [0, 10]
#   * Hiring probability has a range of [0, 100] in units of percentage points
x_exp = np.arange(0, 11, 1)
x_skill = np.arange(0, 11, 1)
x_hiring  = np.arange(0, 101, 1)
```

```python
# Generate fuzzy membership functions
# Experience 
exp_lo = fuzz.trimf(x_exp, abc=[0, 0, 5])
exp_md = fuzz.trimf(x_exp, abc=[0, 5, 10])
exp_hi = fuzz.trimf(x_exp, abc=[5, 10, 10])
# Skill
skill_lo = fuzz.trimf(x_skill, abc=[0, 0, 5])
skill_md = fuzz.trimf(x_skill, abc=[0, 5, 10])
skill_hi = fuzz.trimf(x_skill, abc=[5, 10, 10])
# Hiring probability
hiring_lo = fuzz.trimf(x_hiring, abc=[0, 0, 50])
hiring_md = fuzz.trimf(x_hiring, abc=[0, 50, 100])
hiring_hi = fuzz.trimf(x_hiring, abc=[50, 100, 100])
```

Have a look on the Experience quality membership functions for low, medium and high

```python
print('x_exp',x_exp)
print('exp_lo', exp_lo)
print('exp_md',exp_md)
print('exp_hi',exp_hi)
```


Visualize these universes and membership functions



```python
# Visualize these universes and membership functions
fig, (ax0, ax1, ax2) = plt.subplots(nrows=3, figsize=(8, 9))

ax0.plot(x_exp, exp_lo, 'b', linewidth=1.5, label='Low')
ax0.plot(x_exp, exp_md, 'g', linewidth=1.5, label='Medium')
ax0.plot(x_exp, exp_hi, 'r', linewidth=1.5, label='High')
ax0.set_title('Amount of Experience')
ax0.legend()

ax1.plot(x_skill, skill_lo, 'b', linewidth=1.5, label='Poor')
ax1.plot(x_skill, skill_md, 'g', linewidth=1.5, label='Acceptable')
ax1.plot(x_skill, skill_hi, 'r', linewidth=1.5, label='Amazing')
ax1.set_title('Skill quality')
ax1.legend()

ax2.plot(x_hiring, hiring_lo, 'b', linewidth=1.5, label='Low')
ax2.plot(x_hiring, hiring_md, 'g', linewidth=1.5, label='Medium')
ax2.plot(x_hiring, hiring_hi, 'r', linewidth=1.5, label='High')
ax2.set_title('Hiring Probability')
ax2.legend()

# Turn off top/right axes
for ax in (ax0, ax1, ax2):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```

Question 1
What would the hiring probability be in the following circumstance:

Experience was 2.5
Skill was 9.0

```python
exp_level_lo = fuzz.interp_membership(x_exp, exp_lo, 2.5)
exp_level_md = fuzz.interp_membership(x_exp, exp_md, 2.5)
exp_level_hi = fuzz.interp_membership(x_exp, exp_hi, 2.5)
print(exp_level_lo, exp_level_md, exp_level_hi)
```

```python
skill_level_lo = fuzz.interp_membership(x_skill, skill_lo, 9.0)
skill_level_md = fuzz.interp_membership(x_skill, skill_md, 9.0)
skill_level_hi = fuzz.interp_membership(x_skill, skill_hi, 9.0)
print(skill_level_lo, skill_level_md, skill_level_hi)
```

Rule Evaluation (Inference)
If the experience is low OR the skill is poor, then the tip will be low

```python
import math 

active_rule1 = np.fmax(exp_level_lo, skill_level_lo) # fuzzy-max = Boolean OR
print('active_rule1', active_rule1)
# Now we apply this by clipping the top off the corresponding output
# membership function with `np.fmin`
print(hiring_lo.round(2))
hiring_activation_lo = np.fmin(active_rule1, hiring_lo)  # fuzzy-min = Boolean AND
print(hiring_activation_lo)
```

2. If the experience is medium, then the hiring probability will be medium

```python
# For rule 2 we connect acceptable service to medium tipping
hiring_activation_md = np.fmin(exp_level_md, hiring_md) # fuzzy-min = Boolean AND
print(hiring_md.round(1))
print(hiring_activation_md)
```

3. If the experience is high OR the skill is amazing, then the hiring probability will be high.

```python
# For rule 3 we connect high service OR high food with high tipping
active_rule3 = np.fmax(exp_level_hi, skill_level_hi) # fuzzy-max = Boolean OR
print('active_rule3', active_rule3)
print(hiring_hi)
hiring_activation_hi = np.fmin(active_rule3, hiring_hi)
print(hiring_activation_hi)
```

```python
hiring0 = np.zeros_like(x_hiring)
# Visualize this
fig, ax0 = plt.subplots(figsize=(8, 3))
ax0.fill_between(x=x_hiring, y1=hiring0, y2=hiring_activation_lo, facecolor='b', alpha=0.5)
ax0.plot(x_hiring, hiring_lo, 'b', linewidth=0.5, linestyle='--', )
ax0.fill_between(x_hiring, hiring0, hiring_activation_md, facecolor='g', alpha=0.5)
ax0.plot(x_hiring, hiring_md, 'g', linewidth=0.5, linestyle='--')
ax0.fill_between(x_hiring, hiring0, hiring_activation_hi, facecolor='r', alpha=0.5)
ax0.plot(x_hiring, hiring_hi, 'r', linewidth=0.5, linestyle='--')
ax0.set_title('Output membership activity: Rule3+Rule2+Rule3')

# Turn off top/right axes
for ax in (ax0,):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```

Aggregation (Composition)
Aggregate all three output membership functions together

```python
aggregated = np.fmax(hiring_activation_lo, np.fmax(hiring_activation_md, hiring_activation_hi))
print(aggregated)
```

Defuzzification - 
Calculate defuzzified result using the centroid method

```python
hiring = fuzz.defuzz(x_hiring, aggregated, 'centroid')
print(hiring)
```

```python
tip_activation = fuzz.interp_membership(x_hiring, aggregated, hiring)  # for plot
tip_activation
```
